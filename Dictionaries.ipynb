3. Traversing a Dictionary

```python
myDict = {'Name':'Sanidhya','Age':21,'Branch':'CSE','Address':'Ghaziabad'}

#  Traverse through a dictionary

def traverseDict(dict):
    for key in dict:
        print(key, dict[key])

traverseDict(myDict)
```

4. Searching in a Dictionary

```python
def searchdict(dict,value):
    for key in dict:
        if dict[key]==value:
            return key,value
        
    return 'The value does not exist'

print(searchdict(myDict,21))            
```

```python
myDict = {'Name':'Sanidhya','Age':21,'Branch':'CSE','Address':'Ghaziabad'}
del myDict['Age']
print(myDict)
```
